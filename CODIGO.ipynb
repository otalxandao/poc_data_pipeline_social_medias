## ELT json Votes

```python
import pandas as pd
import json

# Carregar o arquivo JSON como um objeto Python
caminho_arquivo = r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Anexos\votes.json'

with open(caminho_arquivo) as file:
    dados_json = json.load(file)

# Use json_normalize para converter o JSON em um DataFrame
df = pd.json_normalize(dados_json)

# Explodir a coluna 'votes' para expandir os arrays
df = df.explode('votes', ignore_index=True)

# Expandir as colunas 'votes' em colunas separadas ('index' e 'weight')
df[['index', 'weight']] = pd.json_normalize(df['votes'])

# Remover a coluna 'votes' original
df = df.drop(columns='votes')

# Exibindo o DataFrame resultante
pd.set_option('display.max_columns', None)  # Mostrar todas as colunas do DataFrame
pd.set_option('display.expand_frame_repr', False)  # Não truncar o DataFrame na exibição


# Novos nomes das colunas
novos_nomes = {
    '_id.$oid': 'id_voto',
    'user_id.$oid': 'id_user',
    'created_at.$date': 'created_at',
    'updated_at.$date': 'updated_at',
    'post_id.$oid': 'id_post',
    'index': 'index',
    'weight': 'weight'
}

# Renomeando as colunas
df = df.rename(columns=novos_nomes)
df= pd.DataFrame(df)
# Preenchendo os valores NaN na coluna 'index' com 0 (zero)
df['index'].fillna(12345, inplace=True)
df['weight'].fillna('', inplace=True)
# Convertendo a coluna 'index' para o tipo int
df['index'] = df['index'].astype(int)

# Substituindo os valores 12345 por uma string vazia
df['index'] = df['index'].replace(12345, '')

# Criando a coluna 'id_index'
df['id_index'] = df['id_post'] + df['index'].astype(str)
df
df.to_excel(r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Excel\votes.xlsx')
```

# ELT translations

```python
import pandas as pd
import json

caminho_arquivo = r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Anexos\posts.json'

with open(caminho_arquivo) as file:
    seu_json_varias_ocorrencias = json.load(file)
# Lista para armazenar os DataFrames resultantes
dataframes = []

# Loop pelas ocorrências do JSON
for item in seu_json_varias_ocorrencias:
    # Extração do _id
    id_ = item['_id']['$oid']

    # Normalização das traduções
    df_translations = pd.json_normalize(item['translations'])

    # Adicionando a coluna com o _id
    df_translations['_id'] = id_

    # Armazenando o DataFrame resultante na lista
    dataframes.append(df_translations)



# Concatenando todos os DataFrames da lista em um único DataFrame
df_resultante = pd.concat(dataframes, ignore_index=True)

# Exibe o DataFrame resultante
df= pd.DataFrame(df_resultante)

# Novos nomes das colunas
novos_nomes = {
    'title': 'title',
    'description': 'description',
    'language': 'language',
    'created_at.$date': 'created_at',
    'updated_at.$date': 'updated_at',
    '_id': 'id_post',
    'update_at.$date': 'updated_at'
}

# Renomeando as colunas
df = df.rename(columns=novos_nomes)
df = df.drop(columns='options')
df
df.to_excel(r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Excel\translations.xlsx')
```

# ELT options

```python
import pandas as pd
import json

caminho_arquivo = r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Anexos\posts.json'

with open(caminho_arquivo) as file:
    seu_json_varias_ocorrencias = json.load(file)

# Lista para armazenar os DataFrames resultantes
dataframes = []

# Loop pelas ocorrências do JSON
for item in seu_json_varias_ocorrencias:
    # Extração do _id
    id_ = item['_id']['$oid']

    # Loop pelas traduções
    for translation in item['translations']:
        options = translation.get('options', [])
        for index, option in enumerate(options):
            # Adiciona cada opção como uma linha no DataFrame
            dataframes.append({'id_post': id_, 'index': index, 'option': option})

# Cria um DataFrame a partir dos dados coletados
df_resultante = pd.DataFrame(dataframes)


# Exibe o DataFrame resultante
df= pd.DataFrame(df_resultante)


# Criando uma nova coluna com a concatenação das colunas 'id_post' e 'index'
df['id_index'] = df.apply(lambda row: f"{row['id_post']}{row['index']}", axis=1)

# Exibindo o DataFrame resultante com a nova coluna
df
df.to_excel(r'C:\Users\Alexandre Nuevo\OneDrive - NUVETO COMUNICACOES LTDA\Desktop\case\Excel\options.xlsx')

```
